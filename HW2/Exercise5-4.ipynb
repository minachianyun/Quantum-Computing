```python
from qiskit import QuantumCircuit, QuantumRegister
from qiskit.quantum_info import Statevector
qc = QuantumCircuit(3)
qc.x(0)
qc.h(1)
qc.x(2)
qc.h(0)
qc.h(2)
qc.barrier()
state1 = Statevector.from_instruction(qc)
qc.cx(0,2)
qc.cx(1,2)
qc.barrier()
state2 = Statevector.from_instruction(qc)
display(qc.draw('mpl'))
print("The Bloch sphere of the state on the first barrier")
display(state1.draw('bloch'))
print("The Bloch sphere of the state on the second barrier")
display(state2.draw('bloch'))
```
